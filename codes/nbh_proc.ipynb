## Neighborhood processing

Filtering images with PIL

```python
from PIL import Image
from PIL import ImageFilter
from pylab import *
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy.ndimage.filters
import numpy as np
```

```python
im0 = Image.open('C:/Users/Misab/Desktop/Image processing/data/moon.jpg')

figure(figsize=(20,20))

subplot(3,4,1)

plt.imshow(im0, cmap=cm.gray)
plt.title('Original image')

subplot(3,4,2)
im2 = im0.filter(ImageFilter.CONTOUR)
plt.imshow(im2, cmap=cm.gray)
plt.title('Contour')

subplot(3,4,3)
im3 = im0.filter(ImageFilter.DETAIL)
plt.imshow(im3, cmap=cm.gray)
plt.title('Detail')

# laplacian filter

subplot(3,4,4)
im4 = im0.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(im4, cmap=cm.gray)
plt.title('Edge enhancement (laplacian convolutian)')

subplot(3,4,5)

im5 = im0.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(im5, cmap=cm.gray)
plt.title('Edge enhancement More')

subplot(3,4,6)
im6 = im0.filter(ImageFilter.EMBOSS)
plt.imshow(im6, cmap=cm.gray)
plt.title('Emboss')

subplot(3,4,7)
im7 = im0.filter(ImageFilter.FIND_EDGES)
plt.imshow(im7, cmap=cm.gray)
plt.title('Find edges')

subplot(3,4,8)
im8 = im0.filter(ImageFilter.SMOOTH)
plt.imshow(im8, cmap=cm.gray)
plt.title('Low pass filter 1')

subplot(3,4,9)
im9 = im0.filter(ImageFilter.SMOOTH_MORE)
plt.imshow(im9, cmap=cm.gray)
plt.title('Low pass filter 2')

subplot(3,4,10)
im10 = im0.filter(ImageFilter.SHARPEN)
plt.imshow(im10, cmap=cm.gray)
plt.title('Sharpen')

# Custom kernels

size = (3,3)
kernel1 = [1,1,1, 0,0,0, -1,-1,-1]
ker1 = ImageFilter.Kernel(size, kernel1,scale=None, offset=0 )

subplot(3,4,11)
im11 = im0.filter(ker1)
plt.imshow(im11, cmap=cm.gray)
plt.title('Custom filter 1')

kernel2 = [1,0,-1,1,0,-1,0,0,-1]
ker2 = ImageFilter.Kernel(size, kernel2, scale=None, offset=0)
subplot(3,4,12)
im12 = im0.filter(ker2)
plt.imshow(im12, cmap=cm.gray)
plt.title('Custom filter 2')
```
