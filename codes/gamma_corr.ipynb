## Image enhancement techniques

Gamma correction

```python
import numpy as np
from skimage import exposure
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('C:/Users/Misab/Desktop/Image processing/data/profile.jpg')
img = np.array(img)

plt.imshow(img)
```

```python
gamma_corrected = exposure.adjust_gamma(img, 2)

plt.imshow(gamma_corrected)
```

```python
plt.title('Without gamma correction')
plt.hist(img.flatten(), 256, edgecolor = 'black')
```

```python
plt.title('With gamma correction')
plt.hist(gamma_corrected.flatten(), 256, edgecolor = 'black')
```
