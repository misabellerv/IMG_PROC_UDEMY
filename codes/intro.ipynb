Perfoming image binarization with threshold

```python
from matplotlib import cm
from pylab import *
from skimage import data

scanned = data.page()

plt.imshow(scanned, cmap = cm.gray)
```

```python
plt.imshow(scanned)
```

```python
scanned
```

```python
np.shape(scanned)
```

```python
thres = np.zeros(np.shape(scanned)).astype('uint8')

threshold = 150

thres[scanned < threshold] = 0
thres[scanned >= threshold] = 255

plt.imshow(thres)
```

```python
plt.imshow(thres, cmap = cm.gray)
```
