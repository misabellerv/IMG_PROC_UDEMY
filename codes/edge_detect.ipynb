## Edge detection


Sobel edge detection with Skimage

```python
from PIL import Image
from skimage import filters
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
img = np.array(Image.open('C:/Users/Misab/Desktop/Image processing/data/moon.jpg'))
img_filt = filters.sobel(img)

plt.figure(figsize = (15,15))
plt.subplot(1,2,1)
plt.imshow(img, cmap=cm.gray)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_filt,cmap = cm.gray)
plt.title('Sobel edge detection skimage')
```

Sobel edge detection with OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('C:/Users/Misab/Desktop/Image processing/data/moon.jpg')
# or
# img = cv2.imread('C:/Users/Misab/Desktop/Image processing/data/moon.jpg', 0)
# 0 for grayscale

sobel_x = cv2.Sobel(img, -1,1,0,ksize=5)
sobel_y = cv2.Sobel(img,-1,0,1,ksize=5)

plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.imshow(sobel_x, cmap=cm.gray)
plt.title('Sobel edge detection x with opencv')

plt.subplot(1,2,2)
plt.imshow(sobel_y, cmap =cm.gray)
plt.title('Sobel edge detection y with opencv')
```

Laplacian edge detection

```python
img = cv2.imread('C:/Users/Misab/Desktop/Image processing/data/moon.jpg',0)

output = cv2.Laplacian(img,-1)

plt.imshow(output, cmap=cm.gray)
```
